```python
import notebook_setup
from src.datasets import load_tinystories_dataloaders
from src.tokenizers import load_tinystories_tokenizer
from src.training import train
from nanoGPT.model import GPT, GPTConfig
```

```python
tokenizer = load_tinystories_tokenizer()
```

```python
train_dataloader, val_dataloader = load_tinystories_dataloaders(tokenizer)
```

```python
config = GPTConfig(
  block_size=512,
  vocab_size=len(tokenizer),
  n_layer=1,
  n_head=12,
  n_embd=768,
  dropout=0.1,
  ln_setting='pre',
  input_vectors=['w', 'w', 'w']
)

model = GPT(config)
model.name = 'nanoGPT_1L_12H_512E'
```

```python
train_results, eval_results = train(model, train_dataloader, val_dataloader)
```
